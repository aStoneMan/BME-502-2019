```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as stats
```

```python
x = np.linspace(-5,5,100)
norm1 = norm(loc=-0.5)
norm2 = norm(loc=0.5)
```

```python
plt.plot(x,norm1.pdf(x))
plt.plot(x,norm2.pdf(x))
```

```python
n = 100
data1 = norm1.rvs(n)
data2 = norm2.rvs(n)
print(data1.mean())
print(data2.mean())
```

```python
_ = plt.hist(data1, bins=10, density = True)
_ = plt.hist(data2, bins=10, density = True)
plt.plot(x,norm1.pdf(x))
plt.plot(x,norm2.pdf(x))
```

```python
stats.ttest_rel(data1,data2)
```
