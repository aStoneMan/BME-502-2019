```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1.2,4.5,7.8,9.9])
```

```python
np.sin(a)
```

```python
x = np.linspace(0,6*np.pi,200)
s = np.sin(x)
c = np.cos(x)
```

```python
plt.plot(x,s,color="k",label="sin(x)")
plt.plot(x,c,color="r",label="cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
