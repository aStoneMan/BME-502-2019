```python
%matplotlib inline
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
x=np.linspace(-5,5,100)
norm1=norm(loc=-0.5)
norm2=norm(loc=0.5)
```

```python
plt.plot(x,norm1.pdf(x))
plt.plot(x,norm2.pdf(x))
```

```python
n=30
data1=norm1.rvs(n)
data2=norm2.rvs(n)
mean1=data1.mean()
mean2=data2.mean()
std1=data1.std()
std2=data2.std()
std12=np.sqrt((std1**2+std2**2)/2.0)
t=(mean2-mean1)/std12/np.sqrt(2.0/n)
print("mean1: ",mean1, "mean2: ",mean2, "t score: ",t)
plt.hist(data1,normed=True)
plt.hist(data2,normed=True)
```

```python
stats.ttest_rel(data1,data2)
```
